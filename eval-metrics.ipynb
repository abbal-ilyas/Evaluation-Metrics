```python
# Importe les bibliothèques nécessaires
from sklearn import svm
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charge le jeu de données
digits = load_digits()

# Divise les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.30,random_state=4)

# Crée un modèle de classification SVM
clf = svm.SVC(kernel="linear", random_state=4)

# Entraîne le modèle sur l'ensemble d'entraînement
clf.fit(X_train, y_train)

# Prédit les étiquettes des données de test
y_pred = clf.predict(X_test)

# Calcule les mesures de performance du modèle
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred,average="weighted")
recall = recall_score(y_test, y_pred,average="weighted")
f1 = f1_score(y_test, y_pred,average="weighted")
confusion_matrix = confusion_matrix(y_test, y_pred)
cm=confusion_matrix



# Display the results in a table
metrics_df = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1 Score"],
    "Value": [accuracy, precision, recall, f1]
})
print(metrics_df)

# Display the confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```
